### Lists
 - Lists use brackets [] 
 - are indexed with indicies starting at zero (e.g., x[0] is the first element in that list)
 - subset (slice) using this format: list_name[start (inclusive) : end (non-inclusive) : step]. Negative steps work backwards
 - some helpful list operations:
is x in my list?
```python
"fast" in my_list # will return True or False
```
Combining lists together
```python
list_1 = ["Clark", 1, "a"] # 3 elements 
list_2 = ["Bruce", 2, "b"] # 3 elements 
list_3 = list_1 + list_2 # 6 elements: ['Clark', 1, 'a', 'Bruce', 2, 'b']
```
mulitiplication of lists (repeating and appending x times) works the same way
 - you can also count occurances of x (my_list.count(x)), find the index position for the first occurance of x (my_list.index(x)), and sort (my_list.sort()).
 - difference between appending a list (.append) and extending (.extend) a list is largely salient when the thing you're adding is another list. Extend will return one continuous list; Append will return you one list (the original one) and the list you're adding will be added as a single element at the end of the first list. Insert workds similarly to append.
 - Interestingly this is probably all easier to do by redefining the lists:
```python
my_list = [1, 4, 36.3, "G", True]
add_list = ["New1", "New2", "New3"]
my_list[2:2] = add_list #This gives list extension at the 2nd position
# [1, 4, 'New1', 'New2', 'New3', 36.3, 'G', True]
my_list[2] = add_list #This gives list insertion at the 2nd position
# [1, 4, ['New1', 'New2', 'New3'], 'New2', 'New3', 36.3, 'G', True]
```
 - we can remove items from a list by defining their index to be []
```python
my_list = [1, 4, 36.3, "G", True]
my_list[3:4] = [] # removes "G", producing [1, 4, 36.3, True]
```

### Sets
`Sets` are groups of unique and unordered elements.  `Sets` also use curly brackets `{...}` but they should not be confused with `dictionaries`. <br>
Union of A and B is a set of all elements from both sets. A.union(B) <br>
Intersection of A and B is a set of elements that both have in common. A.intersection(B) <br>
Difference of the set B from set A(A - B) is a set of elements that are only in A but not in B. A.difference(B) <br>
Symmetric Difference of A and B is a set of elements in A and B but not in both (excluding the intersection). A.symmetric_difference(B) <br>

### Dictionaries
Dictionaries are the only "mapping" data type in python. Dictionaries are defined by their mapping of "keys", which are unique, to "values", which may not be unique. 
```python
dict1 = {'ten':10, 'nine':9, 'eight':8}
dict2 = dict(ten=10, nine=9, eight=8)
```
Use commands such as:
```python
dict1.values()
```
and 
```python
dict1.keys()
```
to access/loop over components, but don't forget the () at the end! 

To add/modify elements:
```python
dict1['ten'] = "11" # adds new item if the key doesn't already exist
# In this case, it adds the value 11 to the key 'ten'.

{**{1: 'one', 2:'two'}, 3:'three'} # use the double asterix to unpack the nested dict 
```


### Ranges
 - a quick way to make a simple list of numbers
 - ranges use the form range(start (inclusive), end (non-inclusive), step)
 - YOU'RE USUALLY USING A RANGE TO CREATE A LIST. Something like
```python
list(range(5,21,2))
```
which would create a list starting at 5 and increasing in steps of 2 up to 20 (19 really). 

### Loops
 - Often we'll loop over elements in a list
```python
for number in [1, 2, 3, 5, 7, 11]:
    print(f'The square of {number} is {number**2}')
```
 - note the colon that starts the loop and in 4-space indent for all looped code.
 - the two main strategies are to either loop over the indicies or to loop over the elements. 
 - the enumerate function is helpful here as it allows you to do both at the same time:
 ```python
original_list = ['a','b','c']
for idx, i in enumerate(original_list):
    print(f"For index position {idx} the paired value is {i}.".format(idx, i))
```
 - we can also use while loops. They have the form:
 ```python
number = 0
while number<10:
    number+=1
    print(number)
```
List comprehension is using a loop to create a list
```python
my_list = [1, 2, 3, 5, 7, 11]
my_list_squared_2 = [num ** 2 for num in my_list]
```

### Almost will certainly bite me in the ass at some point:
```python
x = [1, 2, 3]
y = x
```
This is going to make y the same as x, so that any time I change x, y will also change. This is a big difference. If we don't want this to happen, we need to use a copy function.
```python
import copy # need this to use deep copy (see below)
x = [1,2,3]
y = x.copy # creates a copy at the first level only
z = copy.deepcopy(x) # creates a copy on all levels (all nested levels)
```

````python
numbers = [6, 9, 3, 1]
sorted(numbers, reverse=True) # The original numbers variable is unchanged because sorted() provides sorted output and does not change the original value in place. The original numbers variable is unchanged because sorted() provides sorted output and does not change the original value in place. 
words = ['banana', 'pie', 'Washington', 'book']
sorted(words, key=len)
sorted(words, key=str.lower) # note that it's the generic "str" not "words"
def reverse_word(word):
    return word[::-1]
sorted(words, key=reverse_word) # sorted on the last letter of the word, using a user-defined function. Nifty.
````

Instead of writing a standalone function, you can use a lambda function defined in the key argument. A lambda is an anonymous function that: <br>
- Must be defined inline
- Doesn’t have a name
- Can’t contain statements
- Will execute just like a function


````python
sorted(words, key=lambda x: x[::-1]) # does the same as the immediately preceeding code. 
````

Sort() versus sorted <br>
Sorted creates (returns) a new list. <br>
Sort() returns nothing, but will modify the existing list.

````python
phrases = ['when in rome', 
    'what goes around comes around', 
    'all is fair in love and war'
    ]
phrases.sort(key=lambda x: x.split()[2][1], reverse=True)
phrases # Note that the original list "phrases" has changed
# Split each phrase into a list of words; 
# Find the third element, or word in this case; 
# Find the second letter in that word

# Sorting dicitonaries by value
xs = {'a': 4, 'b': 3, 'c': 2, 'd': 1}
sorted(xs.items(), key=lambda x: x[1])
````

**Mapping** Processing Iterables Without a Loop (see also: https://realpython.com/python-map-function/)

Since map() is written in C and is highly optimized, its internal implied loop can be more efficient than a regular Python for loop. This is one advantage of using map().

A second advantage of using map() is related to memory consumption. With a for loop, you need to store the whole list in your system’s memory. With map(), you get items on demand, and only one item is in your system’s memory at a given time.

````python
numbers1 = [1, 2, 3] 
numbers2 = [4, 5, 6] 
  
result = map(lambda x, y: x + y, numbers1, numbers2) 
list(result) # note that the result itself is a map, not a list.

with_dots = ["processing..", "...strings", "with....", "..map.."]
list(map(lambda s: s.strip("."), with_dots))

#Mapping to help cleaning text (removing punctuation)
import re
def remove_punctuation(word):
    return re.sub(r'[!?.:;,"()-]', "", word)
text = """Some people, when confronted with a problem, think
"I know, I'll use regular expressions."
Now they have two problems. Jamie Zawinski"""
words = text.split()
words
list(map(remove_punctuation, words))
````

````python
def word_search(doc_list, keyword='fatty'):
    """
    All the descriptive goodness goes in this doc string. This function takes in a list (doc_list)
    and a keyword and kicks out a list identifying the index value of the original list where
    the keyword was found. 
    
    Multi-line docstrings consist of a summary line just like a one-line docstring, 
    followed by a blank line, followed by a more elaborate description. 
    See https://www.python.org/dev/peps/pep-0257/ for best practices
    """
    return_index = []
    try:
        for ind, words in enumerate(doc_list):
            tokens = words.split() # made a new list made up of the words from the 0th string object of documents (word)
            normalize = [token.strip(',.').lower() for token in tokens]
            if keyword.lower() in normalize:
                return_index.append(ind)
    except TypeError: # See other exceptions here: https://docs.python.org/3/library/exceptions.html
        print("Ya gotta use strings, not numbers silly")
    return return_index
                     
doc_list = ["The Learn Python Challenge Casino.", "They bought a casino car", "Casinoville"]
word_search(doc_list, 'casino')
````

### Regular Expressions
Suggestions from [7 Useful Tricks for Python Regex to Learn](https://towardsdatascience.com/7-useful-tricks-for-python-regex-you-should-know-ec20381e22f2#79b2) <br>

When we write a regex pattern, sometimes we have to use slashes or other special characters. Therefore, to avoid that Python interpreter would interpret the string in the wrong way. It is always recommended to use “r-string” when defining a regex pattern.
```python
re.search(r'(\w+)\s+\1', 'abc abc')
```
If we don’t use “r-string” here, the “group” indicator \1 won’t be recognised.<br>

Create regex patterns that are case-insensitive.
```python
re.search(r'[a-zA-Z]+', 'AbCdEfG')
re.search(r'[a-z]+', 'AbCdEfG', re.I) #same as above
```

Make regex statements verbose to better explain/comment the goal.
```python
re.search(r'''
    (\w+)   # Group 1: Match one or more letters, numbers or underscore
    \s+     # Match one or more whitespaces
    \1      # Match the Group 1 whatever it is
''', 'abc   abc', re.VERBOSE)
```
Customize Substitution Behaviour of re.sub()
```python
re.sub(r'\d+', lambda s: '*' * (len(s[0])-3) + s[0][-3:], 'User\'s mobile number is 1234567890')
```

<ins>The Dot Product obeys the following laws:</ins> <br>
$\pmb a \cdot \pmb b = \pmb b \cdot \pmb a$ <br>
$\pmb a \cdot (\pmb b + \pmb c) = \pmb a \cdot \pmb b + \pmb a \cdot \pmb c$ <br>
$(c\pmb a) \cdot \pmb b = c(\pmb a \cdot \pmb b)$ <br>
$\pmb 0 \cdot \pmb a = 0$

````python
a = np.array([1,2,1,2])
b = np.array([-1,2,-1,2])
c = np.array([-1,2,2,-1])
f = np.array([1,-1,1,1])
print(f'Dot product of {a} and {b} is {np.dot(a,b)}.')
print(f'theta for a-f is {math.degrees(math.acos(np.dot(a, f) / (math.sqrt(np.dot(a,a)) * math.sqrt(np.dot(f,f)))))}.')
print(f'theta for b-f is {math.degrees(math.acos(np.dot(b, f) / (math.sqrt(np.dot(b,b) * math.sqrt(np.dot(f,f))))))}.')
print(f'theta for c-f is {math.degrees(math.acos(np.dot(c, f) / (math.sqrt(np.dot(c,c)) * math.sqrt(np.dot(f,f)))))}.')

````

````python
alpha = 2
mat = np.array([[1,1,2],[4,5,6]])
nl = '\n'
print(f'Matrix plus scalar: {nl} {mat} + {alpha} = {nl} {alpha + mat}')
print(f'Matrix times scalar: {nl} {mat} * {alpha} = {nl} {alpha * mat}')

my_vector = np.array([1,2,3])
my_matrix = np.array([[0,1,0], [2,2,1], [3,3,3]])
my_matrix2 = np.array([[4,-1], [4,2]])
my_matrix3 = np.array([[(2/5),(1/5)], [(3/5),(3/5)]])
print(f'Matrix plus vector: {nl} {my_matrix} + {my_vector} = {nl} {my_matrix + my_vector}')
print(f'Matrix times vector: {nl} {my_matrix} * {my_vector} = {nl} {np.dot(my_matrix,my_vector)}')
print(f'Matrix plus matrix: {nl} {my_matrix2} + {my_matrix3} {nl} = {nl} {my_matrix2 + my_matrix3}')
print(f'Matrix times matrix: {nl} {my_matrix2} * {my_matrix3} {nl} = {nl} {np.dot(my_matrix2,my_matrix3)}')
````
